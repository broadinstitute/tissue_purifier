### Get a pretrained classifier

```python
#backbone = torch.hub.load('pytorch/vision:v0.9.0', 'resnet18', pretrained=True)
#backbone.conv1 = torch.nn.Conv2d(9, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
#backbone.fc = torch.nn.Linear(in_features=512, out_features=2, bias=True)
#resnet_classifier = torch.nn.Sequential(backbone, torch.nn.Softmax(dim=-1))

class RandomClassifier(torch.nn.Module):
    def __init__(self, n_classes: int):
        super().__init__()
        self.n_classes = n_classes
        self.activation = torch.nn.Softmax(dim=-1)

    def forward(self, x: torch.Tensor):
        batch_size = x.shape[0]
        y = torch.randn((batch_size, self.n_classes), dtype=x.dtype, device=x.device)
        return self.activation(y)

dummy_classifier = RandomClassifier(n_classes=2)
```

### Run the classifier on a batch of size 1 to check that it runs properly

```python
output = dummy_classifier(rasterized[None,:,:124,:124])
print(output.shape)
print(output[0].detach().cpu().numpy())
```

### Use the classifier on the entire tissue using a overlapping window strategy

```python
input_size = 224

transform_for_tiling = torchvision.transforms.Compose([
    StackTensor(dim=-4),
    RandomGaussianBlur(sigma=(1.0, 1.0)),
    torchvision.transforms.Resize(input_size)
])

output, counter = sparse_image.classify_with_tiling(patch_size=40,
                                                    stride=20,
                                                    classifier=dummy_classifier,
                                                    transform=transform_for_tiling,
                                                    n_element_min=10)
```

```python
plt.imshow(counter[0])
```

```python
av_classifier = torch.where(counter==0, torch.zeros_like(output), output/counter)
```
