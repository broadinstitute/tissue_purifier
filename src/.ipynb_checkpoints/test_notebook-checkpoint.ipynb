# Create a torch Dataset

```python
import lightly 

si_wt2 = SparseImage.from_panda(df=df_wt2, x="x", y="y", category="max_cell_type", pixel_size=10, padding=10)
si_wt3 = SparseImage.from_panda(df=df_wt3, x="x", y="y", category="max_cell_type", pixel_size=10, padding=10)

crop_size = 100

torchdataset_train = SparseDataset(x=[si_wt2.data, si_wt2.data],
                                   y=["wt","aa"],
                                   transform=RandomCropSparseTensor(n_crops=10, crop_size=int(1.5*crop_size)))
tmp = torchdataset_train.__getitem__(0)
print("dataset output two lists -> image_list, label_list", len(tmp), len(tmp[0]), len(tmp[1]))


train_transform = torchvision.transforms.Compose([
    DropoutSparseTensor(dropout=(0.0, 0.4)),
    RasterizeSparseTensor(sigma=(3.0, 3.5)),
    StackDenseTensor(dim=-4),
    torchvision.transforms.RandomAffine(degrees=180, 
                                        scale=(0.75, 1.25), 
                                        shear=2.0, 
                                        interpolation=torchvision.transforms.InterpolationMode.NEAREST, 
                                        fill=0),
    torchvision.transforms.CenterCrop(size=crop_size)
])
```

```python
lightlydataset_train = lightly.data.LightlyDataset.from_torch_dataset(dataset=torchdataset_train)    
    
dataloader_train_simclr = torch.utils.data.DataLoader(
    lightlydataset_train,
    batch_size=lightlydataset_train.__len__(),
    collate_fn=SimClrCollateFn(transform=train_transform),
    shuffle=False)

(imgs0, imgs1), labels, list_filenames = next(iter(dataloader_train_simclr))
```
