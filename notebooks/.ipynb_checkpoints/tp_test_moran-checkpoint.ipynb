### Read in all the csv file

```python
#data_folder = "/home/jupyter/data/slide-seq/original_data/"
data_folder = "../slide-seq-data"

df_wt1 = pd.read_csv(os.path.join(data_folder, "wt_1.csv"))
```

```python
labels_sparse_images = [0]
names_sparse_images = ["wt1"]

sparse_image = tp.data_utils.SparseImage.from_panda(
    df=df_wt1, x="x", y="y", category="cell_type", pixel_size=config["simulation"]["PIXEL_SIZE"], padding=10
)
```

```python
moran = tp.data_utils.SpatialAutocorrelation(modality='moran', n_neighbours=6)
x = moran(sparse_image).sort()[0]
y = [0.01536052, 0.02177229, 0.05939971, 0.09418572, 0.11580584, 0.21362052,
 0.3190746,  0.47539329, 0.52125728]
print(x)
```

```python
import numpy
plt.scatter(x.cpu().numpy(), y)
plt.plot(numpy.linspace(0,0.6,100), numpy.linspace(0,0.6,100))
```
