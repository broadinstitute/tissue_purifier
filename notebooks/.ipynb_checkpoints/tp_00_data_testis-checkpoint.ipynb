### Read in the config file

```python
%matplotlib inline
import torch
import matplotlib.pyplot as plt
from tissue_purifier.data_utils.datamodule import SlideSeqTestisDM
from tissue_purifier.plot_utils.plot_data import show_raw_all_channels, show_raw_one_channel, plot_channel_counts
from tissue_purifier.misc_utils.misc import channel_counter_in_window
```

```python
config_dict_default = SlideSeqTestisDM.get_default_params()
config_dict_default
```

```python
config_dict_v0 = config_dict_default.copy()

zoom_factor = 1 #2 #1 #,2,4

if zoom_factor == 1:
    config_dict_v0["pixel_size"] = 4.0  
    config_dict_v0["n_crops_for_tissue_test"] = 100
    config_dict_v0["n_element_min_for_crop"] = 1000
    config_dict_v0["rasterize_sigma"] = [1.0, 2.0]  
elif zoom_factor == 2:
    ## this zooming in twice as much
    config_dict_v0["pixel_size"] = 2.0  
    config_dict_v0["n_crops_for_tissue_test"] = 400
    config_dict_v0["n_element_min_for_crop"] = 250
    config_dict_v0["rasterize_sigma"] = [2.0, 4.0]  
elif zoom_factor == 4:
    ## this zooming in four time as much
    config_dict_v0["pixel_size"] = 1.0  
    config_dict_v0["n_crops_for_tissue_test"] = 1600
    config_dict_v0["n_element_min_for_crop"] = 100
    config_dict_v0["rasterize_sigma"] = [4.0, 8.0]  
else:
    raise Exception("Zoom factor has the wrong value")
```

```python
dm = SlideSeqTestisDM(**config_dict_v0)
dm.prepare_data()
dm.setup(stage=None)
```

```python
train_dataset = dm.train_dataloader().dataset
print(type(train_dataset))
vars(train_dataset)
```

```python
vars(train_dataset.imgs[0])
```

```python
list_counters = [channel_counter_in_window(img) for img in train_dataset.imgs]
list_f_names = [metadata.f_name for metadata in  train_dataset.metadatas]
cell_names = train_dataset.imgs[0].code_to_cat.values()

print(list_counters)
print(list_f_names)
print(cell_names)
```

## Logger

```python
import neptune.new as neptune
NEPTUNE_PROJECT = "cellarium/tissue-purifier"

exp: neptune.run.Run = neptune.init(project=NEPTUNE_PROJECT,
                                    api_token=os.environ.get('NEPTUNE_API_TOKEN'),
                                    mode="async",
                                    tags=["testis_dataset", "zoom_factor_"+str(zoom_factor)])
```

## Cell counts

```python
fig_cell_count = plot_channel_counts(
    counters=list_counters, 
    dataset_labels=list_f_names, 
    x_labels=cell_names,
    title="Cell counts in 6 slides",
    figsize=(9,9),
)
exp["cell_counts"].log(neptune.types.File.as_image(fig_cell_count))
```

```python
fig_cell_count
```

## Full Images

```python
from tissue_purifier.data_utils.transforms import *

dense_and_raster =torchvision.transforms.Compose([SparseToDense(), Rasterize(sigma=dm.rasterize_sigma)])
list_fig_all_channel = []
list_fig_one_channel = []

for n,(sp_image, f_name) in enumerate(zip(train_dataset.imgs, list_f_names)):
    dense_img = dense_and_raster(sp_image.data)
    fig_all_channel = show_raw_all_channels(dense_img, cmap="tab10", n_col=3, figsize=(12,12), title=f_name, labels=list(cell_names))
    fig_one_channel = show_raw_one_channel(dense_img, cmap='gray', n_col=3, figsize=(10,10), title=f_name, labels=list(cell_names))
    exp["full_sample/all_channels"].log(neptune.types.File.as_image(fig_all_channel))
    exp["full_sample/one_channel"].log(neptune.types.File.as_image(fig_one_channel))
    list_fig_one_channel.append(fig_one_channel)
    list_fig_all_channel.append(fig_all_channel)
```

```python
list_fig_all_channel[0]
```

# Crops

```python
n_iter = 4
n_examples = 5
```

## Validation

```python
val_dataloader = dm.val_dataloader()[0]
list_img, list_label, list_meta = next(iter(val_dataloader))

all_imgs = []
for n in range(n_iter):
    imgs_tmp = dm.trsfm_reference(list_img[:n_examples])
    all_imgs.append(imgs_tmp)
    
imgs_ref = torch.cat(all_imgs, dim=0)

ref_all_ch_fig = show_raw_all_channels(imgs_ref, cmap="tab10", n_col=n_examples, figsize=(10,10), title="Reference crops", labels=list(cell_names))
exp["crops/all_channels"].log(neptune.types.File.as_image(ref_all_ch_fig))
ref_all_ch_fig
```

```python
ref_one_ch_fig = show_raw_one_channel(imgs_ref[0], cmap='gray', n_col=3, figsize=(10,10), title="Reference crops", labels=list(cell_names))
exp["crops/one_channel"].log(neptune.types.File.as_image(ref_one_ch_fig))
ref_one_ch_fig
```

## Local and global

```python
train_dataloader = dm.train_dataloader()
list_img, list_label, list_meta = next(iter(train_dataloader))

all_imgs_local, all_imgs_global = [], []
for n in range(n_iter):
    imgs_tmp_global = dm.trsfm_train_global(list_img[:n_examples])
    imgs_tmp_local = dm.trsfm_train_local(list_img[:n_examples])
    
    all_imgs_local.append(imgs_tmp_local)
    all_imgs_global.append(imgs_tmp_global)
    
imgs_local = torch.cat(all_imgs_local, dim=0)
imgs_global = torch.cat(all_imgs_global, dim=0)
```

### Global

```python
global_all_ch_fig = show_raw_all_channels(imgs_global, cmap="tab10", n_col=n_examples, figsize=(10,10), title="Global crops", labels=list(cell_names))
exp["crops/all_channels"].log(neptune.types.File.as_image(global_all_ch_fig))
global_all_ch_fig
```

```python
global_one_ch_fig = show_raw_one_channel(imgs_global[0], cmap='gray', n_col=3, figsize=(10,10), title="Global crops", labels=list(cell_names))
exp["crops/one_channel"].log(neptune.types.File.as_image(global_one_ch_fig))
global_one_ch_fig
```

### Local

```python
local_all_ch_fig = show_raw_all_channels(imgs_local, cmap="tab10", n_col=n_examples, figsize=(10,10), title="Local crops", labels=list(cell_names))
exp["crops/all_channels"].log(neptune.types.File.as_image(local_all_ch_fig))
local_all_ch_fig
```

```python
local_one_ch_fig = show_raw_one_channel(imgs_local[0], cmap='gray', n_col=3, figsize=(10,10), title="Local crops", labels=list(cell_names))
exp["crops/one_channel"].log(neptune.types.File.as_image(local_one_ch_fig))
local_one_ch_fig
```
