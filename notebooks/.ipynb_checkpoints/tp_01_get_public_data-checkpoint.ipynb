## SparseImages 
Create a sparse image from a panda_dataframe and do few manipulations. \
The data is stored as a torch.sparse_tensor and can be plot using the to_dense() method. \
The pixel_size determines the size of the resulting dense image 

```python
from tissue_purifier.data_utils import SparseImage 

PIXEL_SIZE=20
sparse_image = SparseImage.from_panda(df_wt1, x="x", y="y", category="cell_type", pixel_size=PIXEL_SIZE, padding=10)

print("sparse_image.is_sparse ->", sparse_image.is_sparse)
print("sparse_image.shape ----->", sparse_image.shape)
```

### Visualize the content using to_dense()

```python
from tissue_purifier.plot_utils.plot_data import CELL_TYPE_NAME_MAPPING
labels = list(CELL_TYPE_NAME_MAPPING.values())

tp.plot_utils.show_tensor(sparse_image.to_dense(), 
                          labels=labels,
                          n_col=3, 
                          cmap=plt.cm.inferno,
                          scale_each_channel=True)
```

### Create few random crops and visualize them

Finally we use the random_crops method to generate few random crops. \
The crops are stored as sparse tensor as well. \
They can be plot using to_dense()

```python
crops_list = sparse_image.random_crops(crop_size=100, n_crops=10, n_element_min=100)
```

```python
print("len(crops_list) ->",len(crops_list))
print("type(crops_list[0]) ->",type(crops_list[0]))
print("crops_list[0].is_sparse ->",crops_list[0].is_sparse)
print("crops_list[0].shape ->",crops_list[0].shape)
```

Here we visualize the first channel from each crop

```python
import torch
tmp = torch.stack([crops_list[i].to_dense()[0] for i in range(len(crops_list))], dim=0)
labels = ["ES crop "+str(i) for i in range(len(crops_list))]

tp.plot_utils.show_tensor(tmp, 
                          labels=labels,
                          n_col=5, 
                          scale_each_channel=True, 
                          cmap=plt.cm.inferno,
                          figsize=(12,6))
```
