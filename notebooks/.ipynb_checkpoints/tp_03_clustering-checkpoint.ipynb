## Helper functions

```python
from sklearn.neighbors import NearestNeighbors
import igraph as ig
import leidenalg
import scipy
from typing import Union, Optional, Tuple
import colorcet as cc
import umap
import ipyplot


def make_graph(distances_nn: numpy.ndarray,
               n_neighbors: int,
               metric: str = 'contrastive'):
    
    # Find the k-neighbours of each point
    fitted_nearest_neighbors = NearestNeighbors(n_neighbors=n_neighbors, metric='precomputed').fit(distances_nn)
    neigh_dist_nk, neigh_index_nk = fitted_nearest_neighbors.kneighbors(return_distance=True)
    source_index_nk = torch.arange(neigh_index_nk.shape[0]).view(-1,1).expand(neigh_index_nk.shape).numpy()
    
    # convert a distance into a edge weight
    if metric == 'euclidean':
        weights_nk = 1.0 - neigh_dist_nk / numpy.max(neigh_dist_nk)
    elif metric == 'contrastive':
        weights_nk = 1.0 / neigh_dist_nk
    else:
        raise Exception("Metric not recognized. Received {0}".format(metric))
        
    # Make a sparse matrix
    weights_e = weights_nk.flatten()
    sources_e = source_index_nk.flatten()
    targets_e = neigh_index_nk.flatten()
        
    # Normalize the edges
    w_nn = scipy.sparse.coo_matrix((weights_e, (sources_e, targets_e)), shape=distances_nn.shape)
    degree_n = numpy.asarray(w_nn.sum(-1)).flatten()
    normed_weights_e = weights_e #/ numpy.sqrt(degree_n[sources_e] * degree_n[targets_e])

    return ig.Graph(
        vertex_attrs={"label": numpy.arange(distances_nn.shape[0], dtype=numpy.int64)},
        edges=list(zip(sources_e, targets_e)),
        edge_attrs={"weight": normed_weights_e},
        graph_attrs={"total_edge_weight": normed_weights_e.sum().item(), "total_nodes": distances_nn.shape[0]},
        directed=False)

def compute_partition(graph, resolution_parameter: float):
    partition_type = leidenalg.RBConfigurationVertexPartition
    partition_kwargs = dict(resolution_parameter=resolution_parameter)
    part = leidenalg.find_partition(graph, partition_type, n_iterations=10, **partition_kwargs)
    labels_n = numpy.array(part.membership)
    sizes = numpy.bincount(labels_n)
    print("resolution={0}, # unique_labels={1}, cluster_sizes={2}".format(resolution_parameter, 
                                                                          sizes.shape[0], 
                                                                          sizes))
    return part


def save_to_local_path(list_of_figure: list):
    list_local_path = []
    for n, fig in enumerate(list_of_figure):
        fname = "./tmp_"+str(n)+".png"
        fig.savefig(fname)
        list_local_path.append(fname)
    return list_local_path
```

# What are the embeddings good for?

### Can I predict moran?

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor

x_all = bbone_embeddings
y_all = morans_all
stratify_by = fnames_all

x_train, x_test, y_train, y_test = train_test_split(
    x_all,
    y_all,
    stratify=stratify_by,
    test_size=0.2,
    random_state=1)

mlp_regression = MLPRegressor(
    hidden_layer_sizes=[],
    solver='adam',
    alpha=0.0001,
    batch_size='auto',
    learning_rate='constant',
    learning_rate_init=0.001,
    max_iter=100000,
    shuffle=True,
    random_state=1,
    tol=1E-6,
    verbose=False,
    n_iter_no_change=100,
    early_stopping=False)
mlp_regression.fit(x_train, y_train)

r2_test = mlp_regression.score(x_test, y_test)
r2_train = mlp_regression.score(x_train, y_train)

y_test_predicted = mlp_regression.predict(x_test)
y_train_predicted = mlp_regression.predict(x_train)


fig, axs = plt.subplots(ncols=3, nrows=1, figsize=(24,8))

axs[0].set_title("train dataset")
axs[0].scatter(y_train, y_train_predicted)
axs[0].set_ylabel("predicted")
axs[0].set_xlabel("actual")

axs[1].set_title("test dataset")
axs[1].scatter(y_test, y_test_predicted)
axs[1].set_ylabel("predicted")
axs[1].set_xlabel("actual")

axs[2].set_title("mlp loss")
axs[2].plot(mlp.loss_curve_)
axs[2].set_ylabel("regressor loss")
axs[2].set_xlabel("iteration")
axs[2].set_yscale("log")
print("We can predict moran with good_accuracy", r2_train, r2_test)
```

### Can I predict the tissue labels?

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

x_all = bbone_embeddings
y_all = fnames_all
stratify_by = fnames_all

x_train, x_test, y_train, y_test = train_test_split(
    x_all,
    y_all,
    stratify=fnames_all,
    test_size=0.2,
    random_state=1)

mlp_classification = MLPClassifier(
    hidden_layer_sizes=[],
    solver='adam',
    alpha=0.0001,
    batch_size='auto',
    learning_rate='constant',
    learning_rate_init=0.001,
    max_iter=100000,
    shuffle=True,
    random_state=1,
    tol=1E-6,
    verbose=False,
    n_iter_no_change=100,
    early_stopping=False)

mlp_classification.fit(x_train, y_train)

mean_accuracy_test = mlp_classification.score(x_test, y_test)
mean_accuracy_train = mlp_classification.score(x_train, y_train)

y_test_predicted = mlp_classification.predict(x_test)
y_train_predicted = mlp_classification.predict(x_train)

cm_train = confusion_matrix(y_train, y_train_predicted, labels=mlp_classification.classes_)
cm_test = confusion_matrix(y_test, y_test_predicted, labels=mlp_classification.classes_)

cm_train_disp = ConfusionMatrixDisplay(confusion_matrix=cm_train,
                                       display_labels=mlp_classification.classes_)
cm_test_disp = ConfusionMatrixDisplay(confusion_matrix=cm_test,
                                      display_labels=mlp_classification.classes_)

fig, axs = plt.subplots(ncols=3, nrows=1, figsize=(24,8))

axs[0].set_title("train dataset")
cm_train_disp.plot(ax=axs[0])

axs[1].set_title("test dataset")
cm_test_disp.plot(ax=axs[1])

axs[2].set_title("mlp loss")
axs[2].plot(mlp_classification.loss_curve_)
axs[2].set_ylabel("regressor loss")
axs[2].set_xlabel("iteration")
axs[2].set_yscale("log")
print("We can predict tissue with excellent accuracy", mean_accuracy_train, mean_accuracy_test)
```
