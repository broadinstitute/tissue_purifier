# Unsupervised Clustering 

### These notebook does 5 things:
1. Ingest the ckpt_file
2. For each sparse image, create many patches and embed them.
3. Perform unsupervised clustering of the *patch embeddings* and generate *patch annotations*. 
4. Combine the patch annotations into large images for visualization.
5. Evaluate the image_annotations at the discrete locations of the original data to generate *cell_or_gene annotations*.

# Compute PCA, UMAP, Leiden for all features

```python
for feature_name in feature_names:
    input_features = embeddings_dict[feature_name]

    smart_pca = SmartPca(preprocess_strategy='z_score')
    smart_umap = SmartUmap(n_neighbors=25, preprocess_strategy='raw', n_components=2, min_dist=0.5)

    embeddings_pca = smart_pca.fit_transform(input_features, n_components=0.9)
    embeddings_umap = smart_umap.fit_transform(embeddings_pca)
    umap_graph = smart_umap.get_graph()
    
    smart_leiden = SmartLeiden(graph=umap_graph)
    cluster_labels = smart_leiden.cluster(resolution=0.3)

    embeddings_dict["pca_"+feature_name] = embeddings_pca
    embeddings_dict["umap_"+feature_name] = embeddings_umap
    embeddings_dict["leiden_"+feature_name] = cluster_labels
    
inspect_dict(embeddings_dict)
```

```python
def plot_all_maps(
        input_dict: dict,
        cmap: "matplotlib.colors.ListedColormap" = plt.cm.inferno,
        figsize: Optional[Tuple[float, float]] = None,
        legend: bool = True,
        ) -> List:
    """ Assumes that the input_dictinary has keys starting with:
    umap_, pca_, regress_, classify_, leiden_ """

    maps = []
    
    # Separate keys in 4 disjoint sets
    k_color, k_label, k_umap, k_pca = [], [], [], []
    for k in input_dict.keys():
        if k.startswith("umap"):
            k_umap.append(k)
        elif k.startswith("pca"):
            k_pca.append(k)
        elif k.startswith("classify") or k.startswith("leiden"):
            k_label.append(k)
        elif k.startswith("regress"):
            k_color.append(k)

    # create nested list to plot UMAP
    features, labels, colors = [], [], []
    for key in k_umap:
        for kc in k_color:
            features.append(input_dict[key].cpu().numpy())
            labels.append(None)
            colors.append(input_dict[kc].cpu().numpy())
        for kl in k_label:
            features.append(input_dict[key].cpu().numpy())
            labels.append(input_dict[kl].cpu().numpy())
            colors.append(None)

    # finally plot UMAP
    for f, l, c in zip(features, labels, colors):
        print(f, l, c)
        tmp = plot_embeddings(
            embeddings=f[:,:2],
            labels=l,
            colors=c,
            x_label="UMAP 1",
            y_label="UMAP 2",
            #title="UMAP {0} {1} {2}".format(f, t, title_suffix),
            # cmap=cmap,
            # figsize=figsize,
            # legend=legend,
            )
        maps.append(tmp)
        
    # create nested list to plot PCA
    features, labels, colors = [], [], []
    for key in k_pca:
        for kc in k_color:
            features.append(input_dict[key].cpu().numpy())
            labels.append(None)
            colors.append(input_dict[kc].cpu().numpy())
        for kl in k_label:
            features.append(input_dict[key].cpu().numpy())
            labels.append(input_dict[kl].cpu().numpy())
            colors.append(None)

    # finally plot PCA
    for f, l, c in zip(features, labels, colors):
        print(f, l, c)
        tmp = plot_embeddings(
            embeddings=f[:,:2],
            labels=l,
            colors=c,
            x_label="PCA 1",
            y_label="PCA 2",
            #title="UMAP {0} {1} {2}".format(f, t, title_suffix),
            # cmap=cmap,
            # figsize=figsize,
            # legend=legend,
            )
        maps.append(tmp)

    return maps
```

### Train a simple NearestNeighbour classifier to predict cluster labels

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X = embeddings_umap
y = cluster_labels.astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    stratify=y, 
    test_size=0.5,
    random_state=1,
    shuffle=True,
)

# Train the classifier and evaluate the performance
kn_classifier = KNeighborsClassifier(1)
kn_classifier.fit(X_train, y_train)
score_test = kn_classifier.score(X_test, y_test)
score_train = kn_classifier.score(X_train, y_train)
y_predicted = kn_classifier.predict(X)


# Make a figure
panel_size = 8
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(2*panel_size, panel_size))

axes[0].set_title("Input data")
axes[0].scatter(embeddings_umap[:, 0], embeddings_umap[:, 1], s=1, c=y, cmap=plt.cm.tab10)   
axes[0].set_xlabel("UMAP 1")
axes[0].set_ylabel("UMAP 2")
    
mask = (y_predicted == y)
axes[1].set_title("train, test scores --> ({0:.3}, {1:.3})".format(score_train, score_test))
axes[1].scatter(embeddings_umap[mask, 0], embeddings_umap[mask, 1], s=1, c=y_predicted[mask], cmap=plt.cm.tab10)
axes[1].scatter(embeddings_umap[~mask, 0], embeddings_umap[~mask, 1], s=20, c=y_predicted[~mask], cmap=plt.cm.tab10)
axes[1].set_xlabel("UMAP 1")
axes[1].set_ylabel("UMAP 2")
```
