```python
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support

def train(model, data_loader, optimizer, loss_func):
    model.train()
    for batch_data, batch_labels in data_loader:
        # Batches should come here already from GPU if it's available,
        # TODO: Try to get rid of this check
        if torch.cuda.is_available() and batch_data.device == torch.device("cpu"):
            batch_data = batch_data.cuda()
            batch_labels = batch_labels.cuda()

        net_out = model(batch_data)
        loss = loss_func(net_out, batch_labels.long())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()


def evaluate(model, data_loader):
    model.eval()
    loss_func = torch.nn.CrossEntropyLoss()

    n_correct = 0
    n_wrong = 0
    running_loss = 0.
    counter = 0
    precision = recall = f1_score = 0.
    conf_matrix = np.zeros((2, 2))
    y_trues = []
    y_preds = []
    for batch_data, batch_labels in data_loader:
        with torch.no_grad():
            net_out = model(batch_data)
            y_pred = torch.max(net_out, 1)[1]
            running_loss += loss_func(net_out, batch_labels.long())

            n_correct_batch = (y_pred == batch_labels).sum().item()
            n_wrong_batch = batch_labels.shape[0] - n_correct_batch
            n_correct += n_correct_batch
            n_wrong += n_wrong_batch

            if batch_labels.device != torch.device("cpu"):
                batch_labels = batch_labels.cpu()
                y_pred = y_pred.cpu()

            run_precision, run_recall, run_fscore, _ = precision_recall_fscore_support(
                batch_labels,
                y_pred,
                average="binary"
            )
            y_trues.append(batch_labels)
            y_preds.append(y_pred)

        precision += run_precision
        recall += run_recall
        f1_score += run_fscore
        counter += 1

    y_trues = torch.cat(y_trues)
    y_preds = torch.cat(y_preds)

    tn, fp, fn, tp = confusion_matrix(y_trues, y_preds).ravel()
#     tp, fp, fn, tn = confusion_matrix(y_trues, y_preds).ravel()
    tpr = tp / (tp + fn)
    tnr = tn / (tn + fp)
    mean_loss = running_loss / counter
    accuracy = n_correct / (n_correct + n_wrong)
    precision /= counter
    recall /= counter
    f1_score /= counter

    return {
        "loss": mean_loss,
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1_score": f1_score,
        # Confusion Matrix
        "cm_true_positives": tp,
        "cm_false_positives": fp,
        "cm_false_negatives": fn,
        "cm_true_negatives": tn,
        "cm_true_positive_rate": tpr,
        "cm_true_negative_rate": tnr
    }


def train_eval(model, train_data_loader, eval_data_loader, optimizer, loss_func):
    train(model, train_data_loader, optimizer, loss_func)
    return evaluate(model, eval_data_loader)
```
