```python
NUM_WORKERS = 8
BATCH_SIZE = 128
SEED = 1
MAX_EPOCHS = 100
INPUT_SIZE = 224
NUM_FTRS = 32
PIXEL_SIZE = 4.0
CROP_SIZE = INPUT_SIZE
INPUT_CHANNELS = 9
N_ELEMENT_MIN = 200
N_CROPS_TEST = 150
```

## Raw Data Observation

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(8, 4))
tp.plot_utils.show_raw_data(df_wt1, title="Sample WT 1", ax=ax1, point_size=1.5)
tp.plot_utils.show_raw_data(df_dis1, title="Sample DIS 1", ax=ax2, point_size=1.5)
```

## Data Rasterisation

```python
all_df = [df_wt1, df_wt2, df_wt3, df_dis1, df_dis2, df_dis3]
labels_sparse_images = ["wt1", "wt2", "wt3", "dis1", "dis2", "dis3"]

sparse_images = [
    tp.data_utils.SparseImage.from_panda(
        df, x="x", y="y", category="cell_type", pixel_size=PIXEL_SIZE, padding=10
    )
    for df in all_df
]
```

### Create Test Dataloader

```python
from tissue_purifier.data_utils import transforms as tp_transforms


testdata_transform = torchvision.transforms.Compose(
    [
        tp_transforms.StackTensor(dim=-4),
        tp_transforms.RandomGaussianBlur(sigma=(1.0, 2.0)),
        torchvision.transforms.Resize(INPUT_SIZE),
    ]
)

testdataset = tp.data_utils.SparseDataset(
    x=tp_transforms.RandomCropSparseTensor(
        n_crops=N_CROPS_TEST,
        crop_size=CROP_SIZE,
        n_element_min=N_ELEMENT_MIN,
        test_cache_key="test",
    )(sparse_images),
    y=tp_transforms.Interleave(n_repeat=N_CROPS_TEST)(labels_sparse_images),
)

test_dataloader = tp.data_utils.DataLoaderWithLoad(
    testdataset,
    batch_size=BATCH_SIZE,
    collate_fn=tp_transforms.SpecialCollateFn(
        transform=testdata_transform, simclr_output=False
    ),
    shuffle=False,
)
```

### Create Train Dataloader

```python
from tissue_purifier.data_utils import transforms as tp_transforms


n_crops_for_tissue = int(np.ceil(float(BATCH_SIZE) / len(sparse_images)))

sparse_dataset_train = tp.data_utils.SparseDataset(
    x=sparse_images,
    y=labels_sparse_images,
    transform_x=tp_transforms.RandomCropSparseTensor(
        n_crops=n_crops_for_tissue,
        crop_size=int(1.5 * CROP_SIZE),
        n_element_min=100,
    ),
    transform_y=tp_transforms.Interleave(n_repeat=n_crops_for_tissue),
)

train_transform = torchvision.transforms.Compose(
    [
        tp_transforms.DropoutSparseTensor(dropout=(0.0, 0.4)),
        tp_transforms.StackTensor(dim=-4),
        torchvision.transforms.RandomAffine(
            degrees=180,
            scale=(0.75, 1.25),
            shear=0.0,
            interpolation=torchvision.transforms.InterpolationMode.NEAREST,
            fill=0,
        ),
        tp_transforms.RandomIntensity(factor=(0.7, 1.3)),
        torchvision.transforms.CenterCrop(size=CROP_SIZE),
        tp_transforms.RandomDiagonalCut(offset_bounds=(0.1, 0.9)),
        tp_transforms.RandomGaussianBlur(sigma=(1.0, 5.0)),
        torchvision.transforms.Resize(INPUT_SIZE),
    ]
)

train_dataloader = tp.data_utils.DataLoaderWithLoad(
    sparse_dataset_train,
    batch_size=sparse_dataset_train.__len__(),
    collate_fn=tp_transforms.SpecialCollateFn(
        transform=train_transform, simclr_output=True
    ),
    shuffle=False,
)
```

### Visualize Test Data
